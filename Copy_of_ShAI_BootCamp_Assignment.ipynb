```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

1. **Basic Data Exploration**:

the number of rows and columns in the dataset

```python
x=df.shape
print("Number of rows:", x[0])
print("Number of columns:", x[1])
```

determine the data types of each column

```python
print("Data types of each column: \n",df.dtypes)
```

missing values in each column

```python
print("Missing values in each column: \n",df.isnull().sum())
```

2. **Descriptive Statistics**:

Calculate basic statistics mean, median, mode, minimum, and maximum salary,

```python
print("mean: ",df.TotalPay.mean())
print("median: ",df.TotalPay.median())
print("mode: ",df.TotalPay.mode())
print("minimum: ",df.TotalPay.min())
print("minimum: ",df.TotalPay.max())
```

determine the range of salaries, and find the standard deviation.

```python
print("the range is from : ",df.TotalPay.min()," to ",df.TotalPay.max())
print("standard deviation : ", df.TotalPay.std())
```

3. **Data Cleaning**:

we will drop rows that contain null values in columns (BasePay, OvertimePay, OtherPay) because the null values in these columns are very few compared with the size of the data then deleting the null values is not effict on the data results

```python
df.dropna(subset=['BasePay','OvertimePay','OtherPay'],inplace=True)
```

we will delete the columns (Notes,Status) because all data in these columns are null valuse

```python
df.drop('Notes', axis=1,inplace=True)
df.drop('Status', axis=1,inplace=True)
```

The empty values are not very large, so the column cannot be deleted. Also, the empty values are not small enough to delete the rows without affecting the data, so we will fill these null values.

```python
df['Benefits'].fillna(df['Benefits'].mean(),inplace=True)
```

```python
df.head()
```

```python
print("Missing values in each column: \n",df.isnull().sum())
```

4. **Basic Data Visualization**:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)  # Adjust the number of top departments as needed
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

5. **Grouped Analysis**:

Group the data by 'JobTitle' and calculate summary statistics

```python
grouped_data = df.groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])
print(grouped_data)
```

Group the data by 'Year' and 'JobTitle' and calculate summary statistics

```python
grouped_data = df.groupby(['Year', 'JobTitle'])['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])
print(grouped_data)
```

6. **Simple Correlation Analysis**:

```python
import matplotlib.pyplot as plt

correlation = df['TotalPay'].corr(df['BasePay'])

plt.figure(figsize=(10, 6))
plt.scatter(df['BasePay'], df['TotalPay'], alpha=0.5, color='blue')
plt.title('Scatter Plot of TotalPay vs. BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.text(0.9, 0.1, f'Correlation: {correlation:.2f}', transform=plt.gca().transAxes, color='red', fontsize=12, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt

correlation = df['TotalPay'].corr(df['BasePay'])


plt.figure(figsize=(8, 6))
plt.scatter(df['BasePay'], df['TotalPay'], color='blue', alpha=0.5)
plt.title('Scatter Plot: Total Pay vs Base Pay')
plt.xlabel('Base Pay')
plt.ylabel('Total Pay')
plt.grid(True)
plt.show()

print("Correlation Coefficient between Total Pay and Base Pay:", correlation)
```
